#### train test split
sample 80% of treatments (skipping DMSO)

````python
cpds=[x for x in sf['SM_LINCS_ID'].unique() if x != 'DMSO']
random.seed(123)
traincpds=random.sample(cpds,int(len(cpds)*.9))
testcpds=[x for x in cpds if x not in traincpds]
print 'len cpds:{}, len traincpds:{}, len testcpds:{}'.format(len(cpds),len(traincpds),len(testcpds))
````

#### subsample
For training of replicate-identifying model, select what would be expected to be fairly distinct subset of data.  In particular, reduce 6-point dose response to 2: 10 and 1.1 uM.  Since 3hr time point underrepresented, focus on 24h.

````python
#trainsamp=sf.filter_by(24,'SM_Time').filter_by([10,1.11],'SM_Dose').filter_by(traincpds,'SM_LINCS_ID')
#validsamp=sf.filter_by(24,'SM_Time').filter_by([10,1.11],'SM_Dose').filter_by(validcpds,'SM_LINCS_ID')
#testsamp=sf.filter_by(24,'SM_Time').filter_by([10,1.11],'SM_Dose').filter_by(testcpds,'SM_LINCS_ID')
#print 'train size:{}, valid:{}, test{}'.format(trainsamp.num_rows(),validsamp.num_rows(),testsamp.num_rows())
## rather than make new sframes at this point, add indicator columns to original sframe
````

````python
sf['trainsamp']=sf['SM_LINCS_ID'].apply(lambda x: x in traincpds) * sf['SM_Dose'].apply(lambda x: x in [10,1.11]) * sf['SM_Time'].apply(lambda x: x==24)
print sf.groupby('trainsamp',{'count':sframe.aggregate.COUNT})
sf['testsamp']=sf['SM_LINCS_ID'].apply(lambda x: x in testcpds) * sf['SM_Dose'].apply(lambda x: x in [10,1.11]) * sf['SM_Time'].apply(lambda x: x==24)
print sf.groupby('testsamp',{'count':sframe.aggregate.COUNT})
````

#### probeset columns

````python
import re
pscols=[x for x in sf.column_names() if re.match('^\d+_[srx]?_?at$',x) is not None]
[x for x in sf.column_names() if x not in pscols]
````

Generate hierarchical indicator of treatments indexed as a pseud-float: &lt;cpdid>.&lt;treatid> 

````python
cpdids={x:i for i,x in enumerate(sf['SM_LINCS_ID'].unique())}
sf['cpdid']=sf['SM_LINCS_ID'].apply(lambda x: cpdids[x])
sf['cpdtreatid']=sf['cpdid'].apply(lambda x: str(x))+'.'+sf['treatid'].apply(lambda x: str(x))
sf['cpdtreatid']
````
